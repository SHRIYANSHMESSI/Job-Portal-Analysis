```python
def jobtitle_cleaning(x):
    chars = [':', ';', '(']
    for i in chars:
        if i in x:
            for j in x.split(i):
                if "data" in j.lower():
                    return j.title().strip()
    return x.title().strip()
```

```python
data['job_title'].apply(jobtitle_cleaning).value_counts()
```

    There are many other profiles considering data analyst or data scientist etc but they are with 
    other skills as well. 
    
    Removing those extra keywords and showcasting final result. 

```python
def cleaning_job_title_with_other_skills(x):
    if "-" in x:
        for i in x.split("-"):
            if "data" in x.lower():
                x = i.strip()
                last_val = x.split()[-1]
                if not last_val.strip().isalpha() or "i" in last_val.lower():
                    x = ' '.join(x.split()[:-1])
                    return x
    return x
```

## creating wordcloud of section job description

```python
from wordcloud import WordCloud
```

```python
job_descriptions = data['job_description'].dropna()
for val in job_descriptions:
    val = str(val)
    tokens = val.split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
        text = ' '.join(tokens)+' '
wordcloud = WordCloud(width=1200, height=600, max_words=200, background_color='black').generate(text)
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad = 0)
plt.savefig("static/img/wordcloud.jpg", bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import imageio

def plot_sine_wave(t):
    plt.plot(t, np.sin(t))
    plt.xlabel('Time')
    plt.ylabel('Amplitude')

# Create a list of frames
frames = []
for t in range(100):
    frames.append(plot_sine_wave(t))

# Combine the frames into a GIF
imageio.mimsave('sine_wave.gif', frames, fps=10)
```

```python
import numpy as np
from matplotlib.animation import FuncAnimation
from IPython import display
import matplotlib.pyplot as plt

Figure = plt.figure()

# creating a plot
lines_plotted = plt.plot([]) 

# putting limits on x axis since
# it is a trigonometry function 
# (0,2∏)
line_plotted = lines_plotted[0]

plt.xlim(0,2*np.pi) 

# putting limits on y since it is a
# cosine function
plt.ylim(-1.1,1.1) 

# initialising x from 0 to 2∏
x = np.linspace(0,2*np.pi,100) 

#initially
y = 0

# function takes frame as an input
def AnimationFunction(frame): 

    # setting y according to frame
    # number and + x. It's logic
    y = np.cos(x+2*np.pi*frame/100) 

    # line is set with new values of x and y
    line_plotted.set_data((x, y))

anim_created = FuncAnimation(Figure, AnimationFunction, frames=100, interval=25)    
video = anim_created.to_html5_video()
html = display.HTML(video)
display.display(html) 

# good practice to close the plt object.
plt.close()

anim_created.save("fin.gif")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
column_for_countplot = 'employment_type'

column_for_countplot = 'employment_type'

# Create the GIF
fig, ax = plt.subplots(figsize=(20, 9))  # Adjust the figure size as needed

def update(frame):
    ax.clear()
    sns.barplot(x=data[column_for_countplot].iloc[:frame+1].value_counts().index,
                y=data[column_for_countplot].iloc[:frame+1].value_counts().values,
                ax=ax, palette="rainbow")
    
    ax.set_title(f'Count Plot of {column_for_countplot}')

ani = FuncAnimation(fig, update, frames=len(data), repeat=False)

# Save the animation as a GIF
ani.save('static/img/countplot_animation.gif', writer='pillow', fps=15)


# Display the GIF in Jupyter Notebook
HTML('<img src="countplot_animation.gif">')

```
